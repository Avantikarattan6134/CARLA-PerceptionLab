## 3. Connect to CARLA and Single-Camera Demo
Connect to the CARLA server, spawn an ego vehicle, attach a single RGB camera, and capture a few frames.


```python
import carla
client = carla.Client("127.0.0.1", 2000)
client.set_timeout(20.0)

world = client.get_world()
print("Connected to map:", world.get_map().name)
```

```python
from pathlib import Path
import time

bp = world.get_blueprint_library()
vehicle_bp = bp.filter('vehicle.tesla.model3')[0]
spawn = world.get_map().get_spawn_points()[0]
ego = world.try_spawn_actor(vehicle_bp, spawn)
assert ego is not None, "Failed to spawn vehicle."
ego.set_autopilot(True)

Path("images").mkdir(exist_ok=True)

cam_bp = bp.find('sensor.camera.rgb')
cam_bp.set_attribute('image_size_x', '1280')
cam_bp.set_attribute('image_size_y', '720')
cam_bp.set_attribute('fov', '90')

cam_tf = carla.Transform(carla.Location(x=1.6, z=1.5), carla.Rotation(yaw=0))
cam = world.spawn_actor(cam_bp, cam_tf, attach_to=ego)

def save_img(image):
    image.save_to_disk('images/frame_%06d.jpg' % image.frame)

cam.listen(save_img)
print("Capturing frames for ~10 seconds...")
time.sleep(10)
```

```python
cam.stop()
cam.destroy()
ego.destroy()
print("Cleanup done.")
```

## 4. Multi-Camera Rig (6 Cameras) with Sync Mode
Set up an ego vehicle in autopilot, enable synchronous mode, define a 6-camera rig, record synchronized images, and clean up.


```python
import carla
from pathlib import Path

# blueprint library
bp_lib = world.get_blueprint_library()
vehicle_bp = bp_lib.filter('vehicle.tesla.model3')[0]

# spawn point
spawn_point = world.get_map().get_spawn_points()[0]
ego = world.try_spawn_actor(vehicle_bp, spawn_point)
assert ego is not None, "Failed to spawn ego vehicle."
ego.set_autopilot(True)  # makes it drive automatically

print("✅ Spawned ego vehicle:", ego.id)
```

```python
# Enable sync mode so all sensors tick together
####go synchronous (better alignment)
settings = world.get_settings()
settings.synchronous_mode = True
settings.fixed_delta_seconds = 0.1  # 10 Hz
world.apply_settings(settings)
print("Synchronous mode ON @ 10 Hz")
```

```python
from pathlib import Path
import os, math, time

# Where to save
root = Path("multi_images")
scene = "scene_0002"
env   = "rainy"          # change per weather: clear/rain/fog/night
base  = root / env / scene
base.mkdir(parents=True, exist_ok=True)

# Camera layout (relative to ego vehicle)
CAM_RIG = {
    "CAM_FRONT":       dict(x=1.6,  y=0.0,  z=1.5, yaw=0),
    "CAM_FRONT_RIGHT": dict(x=1.4,  y=0.4,  z=1.5, yaw=45),
    "CAM_FRONT_LEFT":  dict(x=1.4,  y=-0.4, z=1.5, yaw=-45),
    "CAM_BACK":        dict(x=-1.5, y=0.0,  z=1.5, yaw=180),
    "CAM_LEFT":        dict(x=0.0,  y=-0.6, z=1.5, yaw=-90),
    "CAM_RIGHT":       dict(x=0.0,  y=0.6,  z=1.5, yaw=90),
}

# Image settings
W, H = 1280, 720
HFOV_DEG = 90.0

# One folder per camera
for name in CAM_RIG:
    (base / name).mkdir(parents=True, exist_ok=True)

print("Saving to:", base)
```

```python
# Build the camera blueprint
cam_bp = bp_lib.find('sensor.camera.rgb')
cam_bp.set_attribute('image_size_x', str(W))
cam_bp.set_attribute('image_size_y', str(H))
cam_bp.set_attribute('fov', str(HFOV_DEG))

# Spawn + attach
cams = {}
for name, spec in CAM_RIG.items():
    tf = carla.Transform(
        carla.Location(x=spec['x'], y=spec['y'], z=spec['z']),
        carla.Rotation(pitch=0.0, roll=0.0, yaw=spec['yaw'])
    )
    cams[name] = world.spawn_actor(cam_bp, tf, attach_to=ego)

# Simple listeners that save with timestamp (microseconds) so frames line up
listeners = []
def make_listener(name):
    out_dir = base / name
    def _cb(image):
        ts_us = int(image.timestamp * 1e6)
        image.save_to_disk(str(out_dir / f"{ts_us:016d}.jpg"))
    return _cb

for name, cam in cams.items():
    cb = make_listener(name)
    cam.listen(cb)
    listeners.append((cam, cb))

print(f"Spawned and listening on {len(cams)} cameras:", list(cams.keys()))
```

```python
N_SECONDS = 10  # how long to record
ticks = int(N_SECONDS / world.get_settings().fixed_delta_seconds)

print(f"Capturing {N_SECONDS}s …")
for _ in range(ticks):
    world.tick()   # advances all sensors in sync

print("Done. Check:", base)
```

```python
for cam, _ in listeners:
    try:
        cam.stop()
    except:
        pass
    cam.destroy()

# (Optional) return to async mode
s = world.get_settings()
s.synchronous_mode = False
s.fixed_delta_seconds = None
world.apply_settings(s)
print("Cameras destroyed. Synchronous mode OFF.")
```

## 5. Weather Presets & Camera Tuning Utilities
Helper functions and snippets to configure snow/fog/rain presets and cool down camera exposure/white balance.


```python
import carla

def set_weather_snow(world, variant="day"):
    """
    Approximate snowy conditions:
      - heavy cloud cover
      - strong fog (reduced visibility)
      - low sun (flat lighting)
      - zero rain/wetness (keep road dry unless you want slush)
    variant: "day" or "night"
    """
    w = world.get_weather()
    w.cloudiness = 100.0
    w.precipitation = 0.0               # CARLA rain looks like rain, not snow
    w.precipitation_deposits = 0.0
    w.wetness = 0.0                      # set >0 if you want “wet winter” roads
    w.wind_intensity = 60.0

    # fog → key for snowy visibility
    w.fog_density = 80.0
    w.fog_distance = 20.0               # lower = thicker fog near camera
    w.fog_falloff = 2.0

    # cold/flat light
    w.sun_azimuth_angle = 45.0
    w.sun_altitude_angle = 5.0 if variant == "day" else -10.0

    # optional scattering tweaks (if available in your build)
    try:
        w.scattering_intensity = 0.2
        w.mie_scattering_scale = 0.03
        w.rayleigh_scattering_scale = 0.2
    except Exception:
        pass

    world.set_weather(w)
    print(f"Foggy preset applied ({variant}).")
```

```python
def cool_down_camera_bp(cam_bp):
    # If present in your build; safe to ignore if attribute missing
    for k, v in [
        ("exposure_mode", "manual"),
        ("exposure_compensation", "-0.3"),
        ("white_balance_temp", "4500"),   # cooler (default ~6500)
        ("motion_blur_intensity", "0.0"),
    ]:
        if cam_bp.has_attribute(k):
            cam_bp.set_attribute(k, v)
```

```python
import carla

def set_weather_rain(world, variant="day"):
    """
    Realistic rain look:
      - high precipitation + wetness
      - heavy clouds
      - some fog (reduced visibility)
      - lower sun angle for flatter lighting
    variant: "day" or "night"
    """
    w = world.get_weather()
    w.cloudiness = 85.0
    w.precipitation = 90.0            # rain intensity (visual rain)
    w.precipitation_deposits = 80.0   # puddles/wet accumulation
    w.wetness = 95.0                   # reflective wet roads
    w.wind_intensity = 50.0

    w.fog_density = 25.0               # moderate visibility loss
    w.fog_distance = 60.0
    w.fog_falloff = 1.2

    w.sun_azimuth_angle = 45.0
    w.sun_altitude_angle = 15.0 if variant == "day" else -5.0

    # Optional scattering tweaks (ignored if not supported by your build)
    for k, v in [
        ("scattering_intensity", 0.4),
        ("mie_scattering_scale", 0.02),
        ("rayleigh_scattering_scale", 0.3),
    ]:
        try:
            setattr(w, k, v)
        except Exception:
            pass

    world.set_weather(w)
    print(f"Rain preset applied ({variant}).")

# Example:
set_weather_rain(world, variant="day")     # or "night"
```

## 7. Full Scripted Multi-Weather Data Collection Pipeline
A consolidated script-style version of data collection with weather modes and calibration saving.


```python
import carla
import time
import os
import json
import numpy as np

def set_weather(world, mode):
    weather = carla.WeatherParameters.ClearNoon
    if mode == "clear":
        weather = carla.WeatherParameters.ClearNoon
    elif mode == "foggy":
        weather = carla.WeatherParameters(
            cloudiness=20.0, precipitation=0.0, fog_density=80.0, fog_distance=5.0
        )
    elif mode == "rainy":
        weather = carla.WeatherParameters(
            cloudiness=80.0, precipitation=80.0, wetness=100.0
        )
    elif mode == "night":
        weather = carla.WeatherParameters.ClearNight
    world.set_weather(weather)
    print(f"Weather set to: {mode}")

def spawn_camera(world, vehicle, transform, name, out_dir):
    bp = world.get_blueprint_library().find('sensor.camera.rgb')
    bp.set_attribute('image_size_x', '1280')
    bp.set_attribute('image_size_y', '720')
    bp.set_attribute('fov', '90')
    camera = world.spawn_actor(bp, transform, attach_to=vehicle)
    os.makedirs(out_dir, exist_ok=True)

    def save_img(image):
        filename = os.path.join(out_dir, f"{image.frame:06d}.png")
        image.save_to_disk(filename)

    camera.listen(save_img)
    return camera

def save_calibration(camera, file_path):
    """Save intrinsic + extrinsic matrices for PETRv2 converter"""
    cam_transform = camera.get_transform()
    cam_intrinsic = np.identity(3)
    fov = camera.attributes['fov']
    width = int(camera.attributes['image_size_x'])
    height = int(camera.attributes['image_size_y'])
    focal = width / (2.0 * np.tan(float(fov) * np.pi / 360.0))
    cam_intrinsic[0, 0] = cam_intrinsic[1, 1] = focal
    cam_intrinsic[0, 2] = width / 2.0
    cam_intrinsic[1, 2] = height / 2.0

    # Extrinsic (world → camera)
    extrinsic = np.array(cam_transform.get_matrix())

    calib = {"intrinsic": cam_intrinsic.tolist(), "extrinsic": extrinsic.tolist()}
    os.makedirs(os.path.dirname(file_path), exist_ok=True)
    with open(file_path, "w") as f:
        json.dump(calib, f, indent=2)

def main():
    client = carla.Client('localhost', 2000)
    client.set_timeout(10.0)
    world = client.get_world()

    vehicle_bp = world.get_blueprint_library().filter('vehicle.tesla.model3')[0]
    spawn_point = world.get_map().get_spawn_points()[0]
    vehicle = world.spawn_actor(vehicle_bp, spawn_point)

    # Define camera transforms
    cameras = {
        "cam_front": carla.Transform(carla.Location(x=1.6, y=0, z=1.5), carla.Rotation(yaw=0)),
        "cam_front_left": carla.Transform(carla.Location(x=1.4, y=-0.4, z=1.5), carla.Rotation(yaw=-45)),
        "cam_front_right": carla.Transform(carla.Location(x=1.4, y=0.4, z=1.5), carla.Rotation(yaw=45)),
        "cam_back": carla.Transform(carla.Location(x=-1.5, y=0, z=1.5), carla.Rotation(yaw=180)),
        "cam_left": carla.Transform(carla.Location(x=0, y=-0.6, z=1.5), carla.Rotation(yaw=-90)),
        "cam_right": carla.Transform(carla.Location(x=0, y=0.6, z=1.5), carla.Rotation(yaw=90))                           
        
    }

    weather_modes = ["clear", "foggy", "rainy"]
    root_dir = "carla_dataset"

    for mode in weather_modes:
        set_weather(world, mode)
        out_dir = os.path.join(root_dir, mode)
        cameras_list = []
        for name, transform in cameras.items():
            cam = spawn_camera(world, vehicle, transform, name, os.path.join(out_dir, "images", name))
            cameras_list.append((name, cam))

        # Record for a few seconds (you can adjust)
        time.sleep(10)

        # Save calibration JSONs
        for name, cam in cameras_list:
            save_calibration(cam, os.path.join(out_dir, "calib", f"{name}.json"))
            cam.stop()
            cam.destroy()

    vehicle.destroy()
    print("Data collection completed.")

#if _name_ == "_main_":
main()
```
